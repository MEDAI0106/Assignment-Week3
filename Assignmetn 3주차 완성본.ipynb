```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
data = pd.read_csv("C:/Users/com/seaborn-data/owid-covid-data.csv")


dataset_sample = data.loc[:, ['continent', 'date', 'total_cases']]
avg_sample = dataset_sample.groupby('date').total_cases.sum()
avg_sample1 = avg_sample.iloc[:-1]


avg_sample1.plot()
plt.title("Number of the confirmed cases")
plt.xlabel("Date")
plt.ylabel("Count")


plt.show()
```

```python
dataset_sample = data.loc[:, ['continent', 'date', 'total_cases']]
# print(dataset_sample)

dataset_Asia = data[data['continent']=='Asia']
dataset_Asia_sample = dataset_Asia.loc[:, ['date', 'total_cases']]
avg_sample_Asia = dataset_Asia_sample.groupby('date').total_cases.sum()
avg_sample_Asia1 = avg_sample_Asia.iloc[:-1]
# print(avg_sample_Asia1)
avg_sample_Asia1.plot()

dataset_Europe = data[data['continent']=='Europe']
dataset_Europe_sample = dataset_Europe.loc[:, ['date', 'total_cases']]
avg_sample_Europe = dataset_Europe_sample.groupby('date').total_cases.sum()
avg_sample_Europe1 = avg_sample_Europe.iloc[:-1]
# print(avg_sample_Europe1)
avg_sample_Europe1.plot()

dataset_Africa = data[data['continent']=='Africa']
dataset_Africa_sample = dataset_Africa.loc[:, ['date', 'total_cases']]
avg_sample_Africa = dataset_Africa_sample.groupby('date').total_cases.sum()
avg_sample_Africa.plot()

dataset_NorthAmerica = data[data['continent']=='North America']
dataset_NorthAmerica_sample = dataset_NorthAmerica.loc[:, ['date', 'total_cases']]
avg_sample_NorthAmerica = dataset_NorthAmerica_sample.groupby('date').total_cases.sum()
avg_sample_NorthAmerica.plot()

dataset_SouthAmerica = data[data['continent']=='South America']
dataset_SouthAmerica_sample = dataset_SouthAmerica.loc[:, ['date', 'total_cases']]
avg_sample_SouthAmerica = dataset_SouthAmerica_sample.groupby('date').total_cases.sum()
avg_sample_SouthAmerica1 = avg_sample_SouthAmerica.iloc[:-1]
avg_sample_SouthAmerica1.plot()

dataset_Oceania = data[data['continent']=='Oceania']
dataset_Oceania_sample = dataset_Oceania.loc[:, ['date', 'total_cases']]
avg_sample_Oceania = dataset_Oceania_sample.groupby('date').total_cases.sum()
avg_sample_Oceania.plot()

plt.legend(['Asia', 'Europe', 'Africa', 'North America', 'South America', 'Oceania'])
plt.title("Number of the confirmed cases by continent")
plt.xlabel("Date")
plt.ylabel("Count")
```

```python
late_data = data[data['date']> '2021-01-15']
# print(late_data.head())
joint = sns.jointplot(x = 'human_development_index', y = 'total_cases_per_million', 
                      data = late_data, kind = "hex", gridsize = 10)
```

```python
# print(data.info())

data_density = data.loc[:, ['date', 'population_density', 'total_deaths']]
data_density_1 = data_density.groupby('population_density').total_deaths.max().reset_index()
data_density_2 = data_density_1.iloc[:-7, :]
data_density_3 = data_density_2.sort_values(by = 'total_deaths', ascending = False)
data_density_4 = data_density_3.iloc[4:, :]

# print(data_density_4)
ax = plt.subplots()
ax = sns.regplot(x = 'population_density', y = 'total_deaths', data = data_density_4)
ax.set_title('Scatterplot of population density and total deaths')

joint = sns.jointplot(x = 'population_density', y = 'total_deaths', data = data_density_4)
joint.fig.suptitle('Joint Plot of population density and total deaths')
```
